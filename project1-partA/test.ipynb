```python
import pandas as pd

# Load the CSV file with proper handling for headers
data = pd.read_csv('../data_bloomberg/AKAM_data.csv', header=[0, 1], index_col=0, parse_dates=True).copy()
# Inspect the DataFrame to understand the structure

# Identify and drop any extra header rows or misaligned data
# Assuming the extra header row is in index 0
#the first column is date but it doesnt have date as a name
data.columns = data.columns.get_level_values(1)
data.index.name = 'Date'



# Reset the index to get 'Date' as a column if it is not correctly set
data.reset_index(inplace=True)

data.set_index('Date', inplace=True)

# Rename columns if necessary and add Ticker column
data.rename(columns={
    'AKAM US Equity.1': 'PX_LAST',
    'AKAM US Equity.2': 'PX_OPEN',
    'AKAM US Equity.3': 'PX_HIGH',
    'AKAM US Equity.4': 'PX_LOW',
    'AKAM US Equity.5': 'PX_VOLUME',
    'AKAM US Equity': 'CUR_MKT_CAP'
}, inplace=True)

# Add Ticker column
data['Ticker'] = 'AKAM'

# Print the cleaned DataFrame
print(data.head(10))

# Save the cleaned DataFrame to a new CSV file
data.to_csv('data_cleaned_test.csv')
```
